```python
X_train = pc
```

```python
X_test = pd.read_csv('./pc_test.csv')
```

```python
X_test.head()
```

```python
h2o.init()
```

```python
h2train=h2o.H2OFrame(X_train)
h2test=h2o.H2OFrame(X_test)
```

```python
h2train.columns
```

```python
X=[c for c in h2train.columns if c!='Price']

y='Price'
```

```python
automl=H2OAutoML(max_models=10,
                 seed=42,   # random_state
                 max_runtime_secs=300,
                 sort_metric='RMSE')
```

```python
automl.train(x=X,
             y=y,
             training_frame=h2train)
```

```python
print('[INFO] Leader board:')

leader_board=automl.leaderboard

leader_board.head()
```

```python
y_pred=automl.leader.predict(h2test)
```

```python
y_pred[:10]
```

```python
muestra = pd.read_csv('../Proyecto_Kaggle/data/muestra.csv')
```

```python
y_pred_list = h2o.as_list(y_pred)
```

```python
muestra_final = pd.concat([muestra, y_pred_list], axis=1)
```
